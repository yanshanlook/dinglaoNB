斐波那切数列的函数定义

````python
def fib(N):
    F=[]
    a,b=0,1
    while len(F)<N:
        a,b=b,a+b
        F.append(a)
    return F
````

````python
print(fib(10))
````

````python
fib(15)
````

````python
fib(50)
````

````python
fib(100)
````

````python
Runner1={'name':'嘉宁','age':39,'PB':312}
Runner2={'name':'立超','age':38,'PB':438}
Runner3={'name':'乔楠','age':18,'PB':450}
````

````python
def print_PB(Runner):
    print('%s: %s' %(Runner['name'],Runner['PB']))
````

````python
print_PB(Runner1)
print_PB(Runner2)
print_PB(Runner3)
````

````python
class Runner(object):
    def __init__(self,name,age,PB):#构造函数
        self.name=name
        self.age=age
        self.PB=PB
        
    #定义print_PB方法
    def print_PB(self):
        print('%s: %s' %(self.name,self.PB))
        
    #定义running()方法：执行跑步操作
    def running(self):
        print(self.name+" is runing now!")
````

````python
JiaNing=Runner('嘉宁',39,312)
QiaoNan=Runner('乔楠',18,450)
````

````python
JiaNing.print_PB()
QiaoNan.print_PB()
JiaNing.running()
QiaoNan.running()
````

````python
class Runner(object):
    def __init__(self,name,age,PB):#构造函数
        self.name=name
        self.age=age
        self.PB=PB
        
    #定义print_PB方法
    def print_PB(self):
        print('%s: %s' %(self.name,self.PB))
        
    #定义running()方法：执行跑步操作
    def running(self):
        print(self.name+" is runing now!")
````

````python
#定义作者类writer
class Writer(object):
    def __init__(self,name,bookname):
        self.name=name
        self.bookname=bookname
        
    def writting(self):#定义writting()方法
        print(self.name+" is writting "+self.bookname+" book now!")
        
````

````python
#定义写作者类RunningWriter，多继承自两个父类Runner和writer
class RunningWriter(Runner,Writer):
    def __init__(self,name,age,PB,bookname):#构造方法
        Runner.__init__(self,name,age,PB)#调用父类的构造方法
        Writer.__init__(self,name,bookname)
        
    #定义一个驿动写作者类特有的新方法confusing()：纠结如何平衡时间
    def confusing(self):
        print("but How to Balance my Time?")
````

````python
#实例化一个对象
HiFrank=RunningWriter('立超',38,438,'Jupyter')
````

````python
HiFrank.running()
HiFrank.writting()
HiFrank.confusing()
print(HiFrank.name)
````

````python
#定义越野跑者类
class TrailRunner(Runner):
    def climbing(self):
        print(self.name+ "is climbing now !")
    
    #重写running方法，带着登山杖跑
    def running(self):
        print(self.name+" is Running with Trekking poles.")
````

````python
#定义马拉松跑这类
class Marathoner(Runner):
    def running(self):
        print(self.name+" is Steady Running with Pace of 4:50.")
````

````python
JiaNing=Marathoner('嘉宁',39,312)
HiFrank=TrailRunner('立超',38,438)
````

````python
JiaNing.running()
HiFrank.running()
````

````python
#定义一个函数，调用Runner的running方法
def ToRun(Runner):#参数为Runner类的对象
    Runner.running()
    
    
````

````python
#将不同的子类作为ToRun的参数
ToRun(JiaNing)
ToRun(HiFrank)
````
